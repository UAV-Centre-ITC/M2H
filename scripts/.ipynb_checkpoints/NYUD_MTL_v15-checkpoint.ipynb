```python
# Reuired Functions to evalaute the performance
def pixel_accuracy(output, mask):
    with torch.no_grad():
        output = torch.argmax(F.softmax(output, dim=1), dim=1)
        
        # Consider only non-void pixels
        valid = (mask != 0)
        correct = torch.eq(output, mask) & valid
        accuracy = float(correct.sum()) / float(valid.sum())
    return accuracy

def mIoU(pred_mask, mask, smooth=1e-10, n_classes=23):
    with torch.no_grad():
        pred_mask = F.softmax(pred_mask, dim=1)
        pred_mask = torch.argmax(pred_mask, dim=1)
        pred_mask = pred_mask.contiguous().view(-1)
        mask = mask.contiguous().view(-1)

        iou_per_class = []
        for clas in range(1, n_classes): #loop per pixel class : start with 1 to skip void class
            true_class = pred_mask == clas
            true_label = mask == clas

            if true_label.long().sum().item() == 0: #no exist label in this loop
                iou_per_class.append(np.nan)
            else:
                intersect = torch.logical_and(true_class, true_label).sum().float().item()
                union = torch.logical_or(true_class, true_label).sum().float().item()

                iou = (intersect + smooth) / (union +smooth)
                iou_per_class.append(iou)
        return np.nanmean(iou_per_class)
    
def rmse(y_pred, y_true):
    """Compute Root Mean Square Error."""
    return torch.sqrt(torch.mean((y_pred - y_true) ** 2))

def log_rmse(y_pred, y_true):
    """Compute logarithmic Root Mean Square Error."""
    log_pred = torch.log(y_pred + 1)
    log_true = torch.log(y_true + 1)
    return torch.sqrt(torch.mean((log_pred - log_true) ** 2))
```

```python
from torchvision.utils import make_grid
class BerHuLoss(torch.nn.Module):
    def __init__(self, threshold=0.2):
        super(BerHuLoss, self).__init__()
        self.threshold = threshold

    def forward(self, pred, target):
        abs_diff = (pred - target).abs()
        mask = (abs_diff > self.threshold).float()

        l1_loss = abs_diff
        l2_loss = (abs_diff ** 2 + self.threshold ** 2) / (2.0 * self.threshold)

        loss = (1 - mask) * l1_loss + mask * l2_loss

        return loss.mean()
    
def gradient_loss(pred, target):
    dx_pred, dy_pred = torch.gradient(pred, dim=(-2, -1))
    dx_target, dy_target = torch.gradient(target, dim=(-2, -1))

    dx_diff = dx_pred - dx_target
    dy_diff = dy_pred - dy_target

    return torch.mean(dx_diff**2) + torch.mean(dy_diff**2)

class CombinedLoss(torch.nn.Module):
    def __init__(self, threshold=0.2, lambda_gradient=0.5):
        super(CombinedLoss, self).__init__()
        self.berhu_loss = BerHuLoss(threshold)
        self.lambda_gradient = lambda_gradient

    def forward(self, pred, target):
        berhu = self.berhu_loss(pred, target)
        grad_loss = gradient_loss(pred, target)

        return berhu + self.lambda_gradient * grad_loss
    
def post_process_depth(depth_map, method="median", kernel_size=5):
    # Ensure the kernel size is odd
    if kernel_size % 2 == 0:
        raise ValueError("Kernel size must be odd.")

    if method == "median":
        # Padding the tensor
        pad = (kernel_size - 1) // 2
        depth_map_padded = F.pad(depth_map, (pad, pad, pad, pad), mode='reflect')
        # Unfold to apply median filtering
        unfolded = depth_map_padded.unfold(2, kernel_size, 1).unfold(3, kernel_size, 1)
        # Take median along kernel_size x kernel_size window
        depth_map_filtered = unfolded.contiguous().view(unfolded.size()[:-2] + (-1,)).median(dim=-1).values

    else:
        raise ValueError("Unsupported method for post-processing using PyTorch only.")

    return depth_map_filtered

class CosineSimilarityLoss(torch.nn.Module):
    def __init__(self):
        super(CosineSimilarityLoss, self).__init__()
        self.cosine_similarity = torch.nn.CosineSimilarity(dim=1, eps=1e-6)

    def forward(self, predicted, ground_truth):
        # Check if the tensors have the same shape
        if predicted.shape == ground_truth.shape:
            # Ensure the normals are normalized
            predicted_normalized = F.normalize(predicted, p=2, dim=1)
            ground_truth_normalized = F.normalize(ground_truth, p=2, dim=1)
            
            # Compute cosine similarity
            cosine_similarity = self.cosine_similarity(predicted_normalized, ground_truth_normalized)
            
            # This gives a value close to 0 when similarity is high, and up to 2 when it's low
            loss = 1 - cosine_similarity
        else:
            print(f"Shape mismatch: Predicted tensor has shape {predicted.shape} while Ground Truth tensor has shape {ground_truth.shape}")
            loss = torch.full_like(predicted, fill_value=0.95) # create a tensor with the same shape as predicted and fill it with 0.95
        return loss.mean()

###################################################################################################
def train_dinov2(model, w_semseg, w_depth, w_normals, train_loader, val_loader, num_epochs, criterion_segmentation, criterion_depth, optimizer, scheduler, device, checkpoint_path, log_dir): 
    model.train()
    # model.to(device)
    criterion_segmentation.to(device)
    criterion_depth.to(device)
    criterion_surface_normals = CosineSimilarityLoss().to(device)

    logger = SummaryWriter(log_dir=log_dir)

    for epoch in range(num_epochs):
        # Train
        total_segmentation_loss = 0.0
        total_depth_loss = 0.0
        train_miou = 0.0
        train_rmse_depth = 0.0
        total_normals_loss = 0.0
        # pbar = tqdm(total=len(train_loader.dataset), leave=False)
        for i, data in enumerate(train_loader):
            images, labels, depths, normals = data['image'], data['semseg'], data['depth'], data['normals']  
            
            images = images.to(device)
            labels = labels.to(device)
            # print(torch.min(labels), torch.max(labels))
            labels = labels.squeeze(1).long()
            depths = depths.to(device)
            normals = normals.to(device)
            optimizer.zero_grad()

            # Forward pass
            segment_outputs, depth_output, surface_normals, intermediate_d = model(images)
            _, predicted_segmentation = torch.max(segment_outputs, 1)

            loss_segment = criterion_segmentation(segment_outputs, labels)
            loss_depth = criterion_depth(depth_output, depths)  # assuming depth images are 448x448
            rmse_depth = rmse(depth_output, depths)
            loss_normals = criterion_surface_normals(surface_normals, normals)

            # For simplicity, equally weighting both losses. Adjust if needed.
            loss = w_semseg * loss_segment + w_depth * loss_depth + w_normals * loss_normals
            # print(f'{loss.item()}')

            # Backward pass and optimize
            loss.backward()
            optimizer.step()

            total_segmentation_loss += loss_segment.item()
            total_depth_loss += loss_depth.item()
            train_rmse_depth += rmse_depth.item()
            total_normals_loss += loss_normals.item()
            t_miou = mIoU(segment_outputs, labels,n_classes=41)
            train_miou += t_miou
            # pbar.update(images.size(0))
            # pbar.set_postfix({'SegLoss': loss_segment.item(), 
            #                   'DLoss': loss_depth.item(), 
            #                   'NormalsLoss': loss_normals.item(),
            #                   'D-RMSE': rmse_depth.item(),
            #                   'mIoU': t_miou})
            print(f"Batch {i+1}/{len(train_loader)},SegLoss: {loss_segment.item():.4f}, DLoss: {loss_depth.item():.4f},NormalsLoss: {loss_normals.item():.4f}, D-RMSE: {rmse_depth.item():.4f}, mIoU: {t_miou:.4f}", end='\r')

        # Log the scalar values
        info = {'Segmentation Loss': total_segmentation_loss / len(train_loader), 
                'Depth Loss': total_depth_loss / len(train_loader),
                'Normals Loss': total_normals_loss / len(train_loader),
                'Train RMSE Depth': train_rmse_depth / len(train_loader),
                'mIoU': train_miou / len(train_loader)}
                
        for tag, value in info.items():
            logger.add_scalar(tag, value, epoch+1)

        # Validation
        val_loss = 0.0
        val_miou = 0.0
        pixel_accu = 0.0
        val_depth_loss = 0.0
        val_normals_loss = 0.0
        with torch.no_grad():
            for i, data in enumerate(val_loader):
                images, labels, depths, normals = data['image'], data['semseg'], data['depth'], data['normals']  
                images = images.to(device)
                labels = labels.to(device)
                labels = labels.squeeze(1).long()
                depths = depths.to(device)
                normals = normals.to(device)

                segment_outputs, depth_output,surface_normals,intermediate_d = model(images)
                # refined_depth_output = post_process_depth(depth_output)
                loss_segment = criterion_segmentation(segment_outputs, labels)
                loss_depth = rmse(depth_output, depths) # updated to print rmse
                loss_normals = criterion_surface_normals(surface_normals, normals)

                
                loss = loss_segment + loss_depth + loss_normals
                
                val_loss += loss.item()
                val_miou += mIoU(segment_outputs, labels, n_classes=41)
                pixel_accu = pixel_accuracy(segment_outputs, labels)
                val_depth_loss += loss_depth.item()
                val_normals_loss += loss_normals.item()
                
                # Log a single validation output to TensorBoard
                if i == 0:  # only for the first batch in validation
                    # Original RGB Image
                    original_image = make_grid(images[0], normalize=False)
                    logger.add_image('Validation/Original_Image', original_image, epoch+1)

                    # Semantics
                    # Here, we are converting semantic logit outputs to probability scores across 41 classes 
                    # and then determining the class with max probability.
                    pred_grid_semantics = make_grid(segment_outputs[0].max(0)[1].unsqueeze(0) * (255.0 / 40), normalize=True, value_range=(0, 255))
                    true_grid_semantics = make_grid(labels[0].unsqueeze(0) * (255.0 / 40), normalize=True, value_range=(0, 255))

                    logger.add_image('Validation/Predicted_Semantics', pred_grid_semantics, epoch+1)
                    logger.add_image('Validation/True_Semantics', true_grid_semantics, epoch+1)

                    # Depth
                    # Depth
                    pred_grid_depth = make_grid(depth_output[0].clamp(0, 10), normalize=True, value_range=(0, 10))
                    true_grid_depth = make_grid(depths[0].clamp(0, 10), normalize=True, value_range=(0, 10))

                    logger.add_image('Validation/Predicted_Depth', pred_grid_depth, epoch+1)
                    logger.add_image('Validation/True_Depth', true_grid_depth, epoch+1)
                    
                    # Normals
                    pred_grid_normals = make_grid(surface_normals[0], normalize=False)
                    true_grid_normals = make_grid(data['normals'][0].to(device), normalize=False)

                    logger.add_image('Validation/Predicted_Normals', pred_grid_normals, epoch+1)
                    logger.add_image('Validation/True_Normals', true_grid_normals, epoch+1)

                # Remaining validation logging is same as your original code

        # Log the scalar values for validation
        info_val = {
            'val Loss': val_loss / len(val_loader),
            'val RMSE Depth Loss': val_depth_loss / len(val_loader),
            'val Normals Loss': val_normals_loss / len(val_loader),
            'val_mIoU': val_miou / len(val_loader),
            'val_Pixel_accu': pixel_accu / len(val_loader)
        }
        
        for tag, value in info_val.items():
            logger.add_scalar(tag, value, epoch+1)

        # Epoch summary
        print()
        print(f"Epoch {epoch+1}/{num_epochs}, Train Depth Loss: {total_depth_loss/len(train_loader):.4f}, Train mIoU: {train_miou/len(train_loader):.4f}, Val mIoU: {val_miou/len(val_loader):.4f}, Val RMSE: {val_depth_loss / len(val_loader):.4f}")

        # Update the learning rate
        scheduler.step()
        
        # Save model checkpoint
        torch.save(model.state_dict(), f"{checkpoint_path}/model_epoch_{epoch+1}.pt")

    logger.close()

```

```python
class LinearClassifierToken(nn.Module):
    def __init__(self, in_channels, num_chanel=2, tokenW=32, tokenH=32):
        super(LinearClassifierToken, self).__init__()
        self.in_channels=in_channels
        self.W=tokenW
        self.H=tokenH
        self.nc=num_chanel
        self.conv=torch.nn.Conv2d(in_channels,num_chanel,(1,1))
    def forward(self,x):
        return self.conv(x.reshape(-1,self.H,self.W,self.in_channels).permute(0,3,1,2))

class DepthwiseSeparableConv(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size, stride=1, padding=0, bias=True):
        super(DepthwiseSeparableConv, self).__init__()
        
        self.depthwise = nn.Conv2d(in_channels, in_channels, kernel_size, stride, padding, groups=in_channels, bias=False)
        self.pointwise = nn.Conv2d(in_channels, out_channels, 1, 1, 0, bias=bias)

    def forward(self, x):
        x = self.depthwise(x)
        x = self.pointwise(x)
        return x
    
class DepthwiseSeparableDilatedConv(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size=7, dilation=1):
        super(DepthwiseSeparableDilatedConv, self).__init__()
        
        # Calculate padding based on kernel size and dilation
        padding = ((kernel_size - 1) * dilation) // 2

        self.depthwise = nn.Conv2d(in_channels, in_channels, kernel_size=kernel_size, 
                                   padding=padding, groups=in_channels, dilation=dilation)
        self.pointwise = nn.Conv2d(in_channels, out_channels, kernel_size=1)

    def forward(self, x):
        x = self.depthwise(x)
        x = self.pointwise(x)
        return x

    
class DinoV2_448(nn.Module):
    def __init__(self, num_class=1) -> None:
        super().__init__()
        self.dinov2 = torch.hub.load('facebookresearch/dinov2', 'dinov2_vitl14') # large model
        for param in self.dinov2.parameters():
            param.requires_grad = False
        self.n = 2048 
        embed_dim = 1024

        self.num_class = num_class
        self.classlayer_n = LinearClassifierToken(in_channels=embed_dim,num_chanel=self.n,tokenW=32,tokenH=32)
        self.classlayer_n2 = LinearClassifierToken(in_channels=embed_dim,num_chanel=self.n//8,tokenW=32,tokenH=32)
        self.classlayer_n4 = LinearClassifierToken(in_channels=embed_dim,num_chanel=self.n//4,tokenW=32,tokenH=32)
        self.classlayer_n8 = LinearClassifierToken(in_channels=embed_dim,num_chanel=self.n//2,tokenW=32,tokenH=32)
        self.selu = nn.SELU()
        
        self.up = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        self.up_features_448 = nn.Upsample(size=(448, 448), mode='bilinear', align_corners=True)
        self.relu = nn.ReLU(inplace=True)
        
        self.dwsc_nn = DepthwiseSeparableConv(self.n, self.n, kernel_size=7, padding=1)
        self.dwsc_n2 = DepthwiseSeparableConv(self.n + self.n//2, self.n//2, kernel_size=3, padding=1)
        self.dwsc_n4 = DepthwiseSeparableConv(self.n//2 + self.n//4, self.n//4, kernel_size=3, padding=1)
        self.dwsc_n8 = DepthwiseSeparableConv(self.n//4 + self.n//8, self.n//8, kernel_size=3, padding=1)
        
        # dilated 
        self.dwsdc_n16 = DepthwiseSeparableDilatedConv(self.n//8, self.n//16, kernel_size=3, dilation=1)
        self.dwsdc_n32 = DepthwiseSeparableDilatedConv(self.n//16, self.n//32, kernel_size=3, dilation=1)
        self.dwsdc_n64 = DepthwiseSeparableDilatedConv(self.n//32, self.n//64, kernel_size=7, dilation=2)
        self.dwsdc_n128 = DepthwiseSeparableDilatedConv(self.n//64, self.n//128, kernel_size=7, dilation=2)
        
        self.bn_nn = nn.BatchNorm2d(self.n)
        self.bn_n2 = nn.BatchNorm2d(self.n//2)
        self.bn_n4 = nn.BatchNorm2d(self.n//4)
        self.bn_n8 = nn.BatchNorm2d(self.n//8)        
        
        # semantic out layer
        self.semantic_out1 = nn.Conv2d(self.n ,self.num_class,kernel_size=1,bias=True)
        self.semantic_out2 = nn.Conv2d(self.n//4 ,self.num_class,kernel_size=1,bias=True)
        self.semantic_out3 = nn.Conv2d(self.n//8 ,self.num_class,kernel_size=1,bias=True)
        self.semantic_out = nn.Conv2d(self.n//128 ,self.num_class,kernel_size=1,bias=True)
        self.semantic_out_fused = nn.Conv2d(4*self.num_class, self.num_class, kernel_size=1, bias=True)
        
        # depth out layer
        self.depth_out1 = nn.Conv2d(self.n, 1, kernel_size=1, bias=True)
        self.depth_out2 = nn.Conv2d(self.n//4, 1, kernel_size=1, bias=True)
        self.depth_out3 = nn.Conv2d(self.n//8, 1, kernel_size=1, bias=True)
        self.depth_out = nn.Conv2d(self.n//128, 1, kernel_size=1, bias=True)
        self.depth_out_fused = nn.Conv2d(4, 1, kernel_size=1, bias=True)
        
        # normal out
        self.normal_out = nn.Conv2d(self.n//16, 3, kernel_size=1, bias=True)
       
        
        
    def forward(self, x):
        inter_features = []
        # DINOv2 Encoder
        with torch.no_grad():
            out_features = self.dinov2.forward_features(x)['x_norm_patchtokens']
        # print(features.shape)
        all_inter_features = self.dinov2.get_intermediate_layers(x, 12)
        
        # Assuming all_inter_features is a list of features from each layer
        inter_feature_1 = all_inter_features[3]
        inter_feature_2 = all_inter_features[7]
        inter_feature_3 = all_inter_features[11]
        
        inter_x_features1 = self.selu(self.classlayer_n2(inter_feature_1))
        inter_x_features2 = self.selu(self.classlayer_n4(inter_feature_2))
        inter_x_features3 = self.selu(self.classlayer_n8(inter_feature_3))
        
        # Upsample tokenized features to matchign layer resolution
        inter_x_features1_upsampled = F.interpolate(inter_x_features1, size=(224, 224), mode='bilinear', align_corners=True)
        inter_x_features2_upsampled = F.interpolate(inter_x_features2, size=(112, 112), mode='bilinear', align_corners=True)
        inter_x_features3_upsampled = F.interpolate(inter_x_features3, size=(56, 56), mode='bilinear', align_corners=True)
        
        x_features = self.selu(self.classlayer_n(out_features))
        
        # Hard parameter sharing 
        # x_fused = self.fusion_layer(x_features)
        x_f1 = self.dwsc_nn(x_features)
        x_f2 =  self.bn_nn(x_f1)
        x_f2 = self.up(x_f2)
        x_f2 = self.relu(x_f2)
        
        x_f2_fused = torch.cat([x_f2, inter_x_features3_upsampled ], dim=1) 
        x_f2 = self.dwsc_n2(x_f2_fused) # 56*56
        x_f3 = self.bn_n2(x_f2)
        x_f3 = self.relu(x_f3)
        x_f3 = self.up(x_f3)
        
        x_f3_fused = torch.cat([x_f3, inter_x_features2_upsampled ], dim=1) 
        
        

        # Surface Normal Prediction
        x_nf3 = self.dwsc_n4(x_f3_fused) # 112 *112
        x_nf4 = self.bn_n4(x_nf3)
        x_nf4 = self.relu(x_nf4)
        x_nf4 = self.up(x_nf4)
        x_nf4_fused = torch.cat([x_nf4, inter_x_features1_upsampled ], dim=1)
        
        x_n1 = self.dwsc_n8(x_nf4_fused)  # 224 x 224
        x_n2 = self.bn_n8(x_n1)
        x_n2 = self.relu(x_n2)
        x_n2_up = self.up(x_n2)
        x_n3 = self.dwsdc_n16(x_n2_up)  # 448 x 448
        x_n3 = self.relu(x_n3)
        x_n = self.normal_out(x_n3)
        
        
        # depth 
        x_df3 = self.dwsc_n4(x_f3_fused) # 112 *112
        x_df4 = self.bn_n4(x_df3)
        x_df4 = self.relu(x_df4)
        x_df4 = self.up(x_df4)
        x_df4_fused = torch.cat([x_df4, inter_x_features1_upsampled ], dim=1)
        
        x_d1 = self.dwsc_n8(x_df4_fused) #224 * 224
        x_d2 =  self.bn_n8(x_d1)
        x_d2 = self.relu(x_d2)
        x_d2_up = self.up(x_d2)
        x_d3 = self.dwsdc_n16(x_d2_up) 
        x_d3 = self.relu(x_d3)
        x_d4 = self.dwsdc_n32(x_d3) 
        x_d4 = self.relu(x_d4)
        x_d5 = self.dwsdc_n64(x_d4)
        x_d5 = self.relu(x_d5)
        x_d6 = self.dwsdc_n128(x_d5)
        x_d6 = self.relu(x_d6)

        # lets add muli level prediction 
        x_d_out1 = self.depth_out1(x_features)
        x_d_out1 = F.interpolate(x_d_out1 ,size=(448, 448), mode='bilinear', align_corners=True)

        x_d_out2 = self.depth_out2(x_df4)
        x_d_out2 = F.interpolate(x_d_out2 ,size=(448, 448), mode='bilinear', align_corners=True)

        x_d_out3 = self.depth_out3(x_d2)
        x_d_out3 = F.interpolate(x_d_out3 ,size=(448, 448), mode='bilinear', align_corners=True)
        
        x_d_out4 = self.depth_out(x_d6)
        x_d_out_fused = torch.cat([x_d_out1, x_d_out2, x_d_out3, x_d_out4 ], dim=1)
        x_d = self.depth_out_fused(x_d_out_fused)
        
        # semantics
        x_sf3 = self.dwsc_n4(x_f3_fused) # 112 *112
        x_sf4 = self.bn_n4(x_sf3)
        x_sf4 = self.relu(x_sf4)
        x_sf4 = self.up(x_sf4)
        x_sf4_fused = torch.cat([x_sf4, inter_x_features1_upsampled ], dim=1)
        
        x_s1 = self.dwsc_n8(x_sf4_fused) # 224 * 224
        x_s2 =  self.bn_n8(x_s1)
        x_s2 = self.relu(x_s2)
        x_s2_up = self.up(x_s2)
        x_s3 = self.dwsdc_n16(x_s2_up)
        x_s3 = self.relu(x_s3)
        x_s4 = self.dwsdc_n32(x_s3) 
        x_s4 = self.relu(x_s4)
        x_s5 = self.dwsdc_n64(x_s4)
        x_s5 = self.relu(x_s5)
        x_s6 = self.dwsdc_n128(x_s5)
        x_s6 = self.relu(x_s6)

        # multi-level semantics 
        x_s_out1 = self.semantic_out1(x_features)
        x_s_out1 = F.interpolate(x_s_out1 ,size=(448, 448), mode='bilinear', align_corners=True)
        
        x_s_out2 = self.semantic_out2(x_sf4)
        x_s_out2 = F.interpolate(x_s_out2 ,size=(448, 448), mode='bilinear', align_corners=True)
        
        x_s_out3 = self.semantic_out3(x_s2)
        x_s_out3 = F.interpolate(x_s_out1 ,size=(448, 448), mode='bilinear', align_corners=True)
        
        x_s_out4 = self.semantic_out(x_s6)

        x_s_out_fused = torch.cat([x_s_out1, x_s_out2, x_s_out3, x_s_out4 ], dim=1)
        x_s = self.semantic_out_fused(x_s_out_fused)
        

        return x_s, x_d, x_n, x_d_out_fused
```

```python
########## external image

def test_image(model, image_path, device):
    # Load the image
    img_orig = Image.open(image_path).convert('RGB')
    orig_size = img_orig.size  # Save original size for later resizing
    
    # Define the same transformations as your training data
    transform = transforms.Compose([
        # Add any necessary transformations here
        transforms.Resize((448, 448)),  # Assuming you're using this size
        transforms.ToTensor(),
        # Normalize if you normalized your training data
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    
    image = transform(img_orig).unsqueeze(0).to(device)  # Add batch dimension and send to device

    model.eval()

    # Forward pass
    with torch.no_grad():
        segment_outputs, depth_img, normals, intermediate_d = model(image)
        # refined_depth_output = post_process_depth(depth_output)

    # Process outputs (e.g., apply a threshold or argmax for segmentation)
    _, predicted_segmentation = torch.max(segment_outputs, 1)  # Get the most likely label for each pixel

    # Resize outputs to the original size of the image
    resize_transform = transforms.Compose([
        transforms.Resize(orig_size[::-1])  # Use the original size saved earlier
    ])
    
    # For segmentation
    predicted_segmentation_resized = resize_transform(predicted_segmentation.cpu())
    depth_img = resize_transform(depth_img.cpu())
    normals = resize_transform(normals.cpu())
    return predicted_segmentation_resized, depth_img, normals, intermediate_d

# Load trained model
# model_path = "./ckpt_448/model_epoch_30.pt"  
# model = DinoV2_448(num_class=41)
# model.load_state_dict(torch.load(model_path))
# model.to("cuda")  # or "cpu" if you're not using a GPU

# Test model on a random image
image_path = "./image.png"  

predicted_segmentation_img, depth_img, normals, intermediate_d = test_image(model, image_path, "cuda")


fig, axs = plt.subplots(1, 4, figsize=(16, 9), constrained_layout=True)

img_orig = Image.open(image_path)
axs[0].imshow(img_orig)
axs[0].grid(False)

# axs[1,1].imshow(intermediate_d[0].squeeze().cpu().detach().numpy())
# axs[1,1].grid(False)

axs[1].imshow(predicted_segmentation_img.squeeze().cpu().detach().numpy())
axs[1].grid(False)

axs[2].imshow(depth_img.squeeze().cpu().detach().numpy())
axs[2].grid(False)

axs[3].imshow(normals.squeeze().permute(1,2,0).cpu().detach().numpy())
axs[3].grid(False)
```
